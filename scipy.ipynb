```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal as mvn

cov = np.array([[1,0.8],[0.8,5]])
nu = np.array([1,2])
#to get random variable samples
r= mvn.rvs(mean=nu,cov=cov,size=1000)
plt.scatter(r[:,0],r[:,1])
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x= np.linspace(1,100,1000)
y=np.sin(x)+np.sin(3*x)+np.sin(5*x)
plt.plot(y)
plt.show()
Y=np.fft.fft(y)
# fast fourier transforms gives the complex numbers
# we need to find magnitude before ploting

plt.plot(np.abs(Y))
plt.show()

print ('frequencies:',2*np.pi*16/100)
print ('frequencies:',2*np.pi*48/100)
```
